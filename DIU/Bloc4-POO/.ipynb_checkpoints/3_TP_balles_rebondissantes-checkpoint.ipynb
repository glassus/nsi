````python
import pygame, sys
import time
from pygame.locals import *

ecran = (800,600)
taille = 20

pygame.display.init()
fenetre = pygame.display.set_mode(ecran)
fenetre.fill([0,0,0])

x = 300
y = 200
dx = 4
dy = -3
couleur = (45,170,250)

while True :
    fenetre.fill([0,0,0])
    pygame.draw.circle(fenetre, couleur, (x,y), taille)
    
    x += dx
    y += dy
    
    
    # rebonds sur les murs
    if x < taille or x > ecran[0] - taille :
        dx = -dx
        
    if y < taille or y > ecran[1] - taille :
        dy = -dy
    
    pygame.display.update()
    
    # routine pour pouvoir fermer Â«proprementÂ» la fenÃªtre Pygame
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.display.quit()
            sys.exit()
    
    
    time.sleep(0.01)
````

## 2. La POO Ã  la rescousse : crÃ©ation d'une classe Balle

### 2.1 la classe Balle
L'objectif est que la mÃ©thode constructeur dote chaque nouvelle balle de valeurs alÃ©atoires : abscisse, ordonnÃ©e, vitesse, couleur...  
CrÃ©ez cette classe et instanciez une balle, qui devra bouger et rebondir.

> *Remarque : la liste des attributs et des mÃ©thodes de la classe est Ã©laborÃ©e conjointement avec les Ã©lÃ¨ves.*


*Correction possible*

````python
import pygame, sys
import time
from pygame.locals import *
from random import randint

clock = pygame.time.Clock()

ecran = (640,480)
FPS = 30

taille = 20

pygame.display.init()
fenetre = pygame.display.set_mode(ecran)
fenetre.fill([0,0,0])

class Balle :
    def __init__(self):
        self.x = randint(0,ecran[0])
        self.y = randint(0,ecran[1])       
        self.dx = randint(3,7)
        self.dy = randint(-7,-3)
        self.color = (randint(0,255),
                      randint(0,255),
                      randint(0,255))
        self.taille = taille
        
    def move(self):
        # mouvement de la balle self
        self.x += self.dx
        self.y += self.dy
        
        # rebond de la balle self
        if self.x < self.taille or self.x > ecran[0]-self.taille :
            self.dx = -self.dx
        if self.y < self.taille or self.y > ecran[1]-self.taille :
            self.dy = -self.dy
            
  
    def draw(self):
        pygame.draw.circle(fenetre,self.color, (self.x,self.y), self.taille)

b = Balle()
        
while True :

    fenetre.fill([0,0,0])
    b.move()
    b.draw() 
    

    
    pygame.display.flip()
    
    # routine pour pouvoir fermer Â«proprementÂ» la fenÃªtre Pygame
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.display.quit()
            sys.exit()
    
    clock.tick(FPS)
````

### 2.2 Une liste de balles
L'objectif est d'instancier autant de balles que l'on veut, stockÃ©es dans une liste appelÃ©e ```liste_balles```.

````python
import pygame, sys
import time
from pygame.locals import *
from random import randint

clock = pygame.time.Clock()

ecran = (640,480)
FPS = 30

taille = 20
nombre_balles = 10


pygame.display.init()
fenetre = pygame.display.set_mode(ecran)
fenetre.fill([0,0,0])

class Balle :
    def __init__(self):
        self.x = randint(0,ecran[0])
        self.y = randint(0,ecran[1])       
        self.dx = randint(-5,5)
        self.dy = randint(-5,5)
        self.color = (randint(0,255),
                      randint(0,255),
                      randint(0,255))
        self.taille = taille
        
    def move(self):
        # mouvement de la balle self
        self.x += self.dx
        self.y += self.dy
        
        # rebond de la balle self
        if self.x < self.taille or self.x > ecran[0]-self.taille :
            self.dx = -self.dx
        if self.y < self.taille or self.y > ecran[1]-self.taille :
            self.dy = -self.dy            
        
        
    
    def draw(self):
        pygame.draw.circle(fenetre,self.color,
                           (self.x,self.y),
                           self.taille)

liste_balles = []
for k in range(nombre_balles):
    b = Balle()
    liste_balles.append(b)  
        
while True :

    fenetre.fill([0,0,0])
    for b in liste_balles :
        b.move()
        b.draw() 
    

    
    pygame.display.flip()
    
    # routine pour pouvoir fermer Â«proprementÂ» la fenÃªtre Pygame
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.display.quit()
            sys.exit()
    
    clock.tick(FPS)
````

### 2.3 Collision de toutes les balles
L'objectif est de dÃ©tecter et de gÃ©rer la collision de toutes les balles entre elles, par une modification de la mÃ©thode ```move()``` propre Ã  chaque balle.

> *Remarque : une aide peut Ãªtre apportÃ©e oralement, en expliquant qu'il faut dÃ©tecter la collision de la balle ```self``` avec chaque balle ```b``` de Â ```liste_balles```.*

*Correction possible :*

````python
import pygame, sys
import time
from pygame.locals import *
from random import randint

clock = pygame.time.Clock()

ecran = (640,480)
FPS = 30

taille = 20
nombre_balles = 10


pygame.display.init()
fenetre = pygame.display.set_mode(ecran)
fenetre.fill([0,0,0])

class Balle :
    def __init__(self):
        self.x = randint(0,ecran[0])
        self.y = randint(0,ecran[1])       
        self.dx = randint(-5,5)
        self.dy = randint(-5,5)
        self.color = (randint(0,255),
                      randint(0,255),
                      randint(0,255))
        self.taille = taille
        
    def move(self):
        # mouvement de la balle self
        self.x += self.dx
        self.y += self.dy
        
        # rebond de la balle self
        if self.x < self.taille or self.x > ecran[0]-self.taille :
            self.dx = -self.dx
        if self.y < self.taille or self.y > ecran[1]-self.taille :
            self.dy = -self.dy
            
        # collision de la balle self avec les autres balles
        for b in liste_balles :
            if ((b.x-self.x)**2 + (b.y-self.y)**2)**0.5 < self.taille + b.taille :
                # on Ã©change les dx
                b.dx, self.dx = self.dx, b.dx
                # on Ã©change les dy
                b.dy, self.dy = self.dy, b.dy              
        
        
    
    def draw(self):
        pygame.draw.circle(fenetre,self.color,
                           (self.x,self.y),
                           self.taille)

liste_balles = []
for k in range(nombre_balles):
    b = Balle()
    liste_balles.append(b)  
        
while True :

    fenetre.fill([0,0,0])
    for b in liste_balles :
        b.move()
        b.draw() 
    

    
    pygame.display.flip()
    
    # routine pour pouvoir fermer Â«proprementÂ» la fenÃªtre Pygame
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.display.quit()
            sys.exit()
    
    clock.tick(FPS)
````

## 3. Pour aller plus loin : Ã©pidÃ©miologie
Principe : en rajoutant un attribut ```malade``` Ã  chaque balle et en dÃ©cidant qu'une collision entraÃ®ne une contamination, on veut observer la vitesse de propagation de la maladie.

### 3.1 Rajout d'un attribut  ```malade```

````python
import pygame, sys
import time
from pygame.locals import *
from random import randint

clock = pygame.time.Clock()

nb_balles = 30

ecran = (640,480)
FPS = 30

taille = 10

pygame.display.init()
fenetre = pygame.display.set_mode(ecran)
fenetre.fill([0,0,0])

class Balle :
    def __init__(self):
        self.x = randint(0,ecran[0])
        self.y = randint(0,ecran[1])       
        self.dx = randint(-5,5)
        self.dy = randint(-5,5)
        self.color = (255,255,255)
        self.malade = False
        self.taille = taille
        
    def move(self):
        # mouvement de la balle self
        self.x += self.dx
        self.y += self.dy
        
        # rebond de la balle self
        if self.x < self.taille or self.x > ecran[0]-self.taille :
            self.dx = -self.dx
        if self.y < self.taille or self.y > ecran[1]-self.taille :
            self.dy = -self.dy
            
        # collision de la balle self avec les autres balles
        for b in liste_balles :
            if ((b.x-self.x)**2 + (b.y-self.y)**2)**0.5 < self.taille + b.taille :
                # on Ã©change les dx
                b.dx, self.dx = self.dx, b.dx
                # on Ã©change les dy
                b.dy, self.dy = self.dy, b.dy              
        
        
    
    def draw(self):
        if self.malade == True :
            self.color = (255,0,0)
            
        pygame.draw.circle(fenetre,self.color,
                           (self.x,self.y),
                           self.taille)

liste_balles = []
for k in range(nb_balles):
    b = Balle()
    liste_balles.append(b)  

liste_balles[0].malade = True    
    
    
while True :

    fenetre.fill([0,0,0])
    for b in liste_balles :
        b.move()
        b.draw() 
    

    
    pygame.display.flip()
    
    # routine pour pouvoir fermer Â«proprementÂ» la fenÃªtre Pygame
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.display.quit()
            sys.exit()
    
    clock.tick(FPS)
````

### 3.2 Rajout d'une contamination en cas de collision

````python
import pygame, sys
import time
from pygame.locals import *
from random import randint

clock = pygame.time.Clock()

nb_balles = 30

ecran = (640,480)
FPS = 30

taille = 10

pygame.display.init()
fenetre = pygame.display.set_mode(ecran)
fenetre.fill([0,0,0])

class Balle :
    def __init__(self):
        self.x = randint(0,ecran[0])
        self.y = randint(0,ecran[1])       
        self.dx = randint(-5,5)
        self.dy = randint(-5,5)
        self.color = (255,255,255)
        self.malade = False
        self.taille = taille
        
    def move(self):
        # mouvement de la balle self
        self.x += self.dx
        self.y += self.dy
        
        # rebond de la balle self
        if self.x < self.taille or self.x > ecran[0]-self.taille :
            self.dx = -self.dx
        if self.y < self.taille or self.y > ecran[1]-self.taille :
            self.dy = -self.dy
            
        # collision de la balle self avec les autres balles
        for b in liste_balles :
            if ((b.x-self.x)**2 + (b.y-self.y)**2)**0.5 < self.taille + b.taille :
                # on Ã©change les dx
                b.dx, self.dx = self.dx, b.dx
                # on Ã©change les dy
                b.dy, self.dy = self.dy, b.dy 
                # contamination Ã©ventuelle
                contamine(b,self)
 
    def contamine(self, b) :
        if self.malade == True or b.malade == True :
            self.malade = True
            b.malade = True
                
    
    def draw(self):
        if self.malade == True :
            self.color = (255,0,0)
            
        pygame.draw.circle(fenetre,self.color,
                           (self.x,self.y),
                           self.taille)


        
        
        
liste_balles = []
for k in range(nb_balles):
    b = Balle()
    liste_balles.append(b)  

liste_balles[0].malade = True    
    
    
while True :

    fenetre.fill([0,0,0])
    for b in liste_balles :
        b.move()
        b.draw() 
    

    
    pygame.display.flip()
    
    # routine pour pouvoir fermer Â«proprementÂ» la fenÃªtre Pygame
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.display.quit()
            sys.exit()
    
    clock.tick(FPS)
````

### 3.3 Ã‰tude de la vitesse de contamination

````python
import pygame, sys
import time
from pygame.locals import *
from random import randint

%matplotlib inline
import matplotlib.pyplot as plt 


clock = pygame.time.Clock()

nb_balles = 40
duree_simu = 15


ecran = (640,480)
FPS = 25

taille = 10

pygame.display.init()
fenetre = pygame.display.set_mode(ecran)
fenetre.fill([0,0,0])

class Balle :
    def __init__(self):
        self.x = randint(0,ecran[0])
        self.y = randint(0,ecran[1])       
        self.dx = randint(-5,5)
        self.dy = randint(-5,5)
        self.color = (255,255,255)
        self.malade = False
        self.taille = taille
        
    def move(self):
        # mouvement de la balle self
        self.x += self.dx
        self.y += self.dy
        
        # rebond de la balle self
        if self.x < self.taille or self.x > ecran[0]-self.taille :
            self.dx = -self.dx
        if self.y < self.taille or self.y > ecran[1]-self.taille :
            self.dy = -self.dy
            
        # collision de la balle self avec les autres balles
        for b in liste_balles :
            if ((b.x-self.x)**2 + (b.y-self.y)**2)**0.5 < self.taille + b.taille :
                # on Ã©change les dx
                b.dx, self.dx = self.dx, b.dx
                # on Ã©change les dy
                b.dy, self.dy = self.dy, b.dy 
                # contamination Ã©ventuelle
                contamine(b,self)
        
        
    
    def draw(self):
        if self.malade == True :
            self.color = (255,0,0)
            
        pygame.draw.circle(fenetre,self.color,
                           (self.x,self.y),
                           self.taille)

    def contamine(self, b) :
        if self.malade == True or b.malade == True :
            self.malade = True
            b.malade = True
                
        
        
liste_balles = []
for k in range(nb_balles):
    b = Balle()
    liste_balles.append(b)  

liste_balles[0].malade = True    

def nombre_malades():
    s = 0
    for b in liste_balles :
        if b.malade == True :
            s += 1
    return s

liste_nb_malades = []

c = 0

while c < FPS*duree_simu :
    c += 1
    
    fenetre.fill([0,0,0])
    for b in liste_balles :
        b.move()
        b.draw() 
    
    if c % FPS == 0 : # chaque seconde
        liste_nb_malades.append(nombre_malades())   
    
    
    
    pygame.display.flip()
     
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.display.quit()
            sys.exit()
    
    clock.tick(FPS)
    
    
pygame.display.quit()


x=[k for k in range(duree_simu)]
plt.xlabel('temps en secondes')
plt.ylabel('nombre de balles contaminÃ©es')
plt.title('Ã‰tude du nombre de balles contaminÃ©es en ' + str(duree_simu) + " secondes")

plt.plot(x,liste_nb_malades, marker = 'o', c = 'blue',  markersize = 10)
````

### 3.4 Un peu d'espoir (pour les plus courageux)

Apporter trois modifications Ã  la classe Balle :  
- Une chance sur 2 de contamination par contact.
- Un sujet malade guÃ©rit au bout d'un certain temps s'il n'entre pas en contact avec d'autres malades.  
- Un sujet guÃ©ri est immunisÃ© et se colore en vert.


````python
import pygame, sys
from pygame.locals import *
from random import randint

clock = pygame.time.Clock()

nb_balles = 30

ecran = (640,480)
FPS = 30

taille = 10

pygame.display.init()
fenetre = pygame.display.set_mode(ecran)
fenetre.fill([0,0,0])

class Balle :
    def __init__(self):
        self.x = randint(0,ecran[0])
        self.y = randint(0,ecran[1])       
        self.dx = randint(-5,5)
        self.dy = randint(-5,5)
        self.color = (255,255,255)
        self.malade = 0
        self.duree = 0
        self.taille = taille
        
    def move(self):
        # mouvement de la balle self
        self.x += self.dx
        self.y += self.dy

        # rebond de la balle self
        if self.x < self.taille or self.x > ecran[0]-self.taille :
            self.dx = -self.dx
        if self.y < self.taille or self.y > ecran[1]-self.taille :
            self.dy = -self.dy

        # collision de la balle self avec les autres balles
        for b in liste_balles :
            if ((b.x-self.x)**2 + (b.y-self.y)**2)**0.5 < self.taille + b.taille :
                # on Ã©change les dx
                b.dx, self.dx = self.dx, b.dx
                # on Ã©change les dy
                b.dy, self.dy = self.dy, b.dy 
                # contamination Ã©ventuelle
                self.contamine(b)


    def draw(self):
        if self.malade == 1  :
            self.color = (255,0,0)

        pygame.draw.circle(fenetre,self.color,(self.x,self.y),self.taille)
        
    def contamine(self, b):
        proba = randint(0,1)
        if proba == 0 and (self.malade != 2 and b.malade != 2):

            if (b.malade == 1 or self.malade == 1 ):
                b.malade = 1
                self.malade = 1
                self.duree = pygame.time.get_ticks()
                b.duree = pygame.time.get_ticks()


    def guerison(self):
        if self.malade == 1 and (pygame.time.get_ticks() - self.duree) > 200 :
            self.malade = 2
            self.duree = 0
            self.color = (0,255,0)

liste_balles = []
for k in range(nb_balles):
    b = Balle()
    liste_balles.append(b)
           
for i in range(5):           
    liste_balles[i].malade = 1   

while True :

    fenetre.fill([0,0,0])
    for b in liste_balles :
        b.draw()
        b.move()
        b.guerison()
             
   
    pygame.display.flip()

    # routine pour pouvoir fermer Â«proprementÂ» la fenÃªtre Pygame
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.display.quit()
            sys.exit()

    clock.tick(FPS)
````
